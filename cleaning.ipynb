This is random forest example for the kaggle dataset on breast cancer malignancy

https://www.kaggle.com/datasets/saurabhbadole/breast-cancer-wisconsin-state/data

```python
import pandas as pd
```

```python
columns = ["Sample code number", "Clump Thickness",
          "Uniformity of Cell Size", "Uniformity of Cell Shape",
          "Marginal Adhesion", "Single Epithelial Cell Size",
          "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli",
          "Mitoses", "Class"]
```

```python
df = pd.read_csv('breast-cancer-wisconsin.data', sep=',', names=columns)
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
df['Bare Nuclei'].replace('?', 0, inplace=True)
print(df[df['Bare Nuclei'] == '?'].count())
df['Bare Nuclei'] = df['Bare Nuclei'].astype(int)
```

```python
sns.pairplot(df, vars=columns[1:-1], hue=columns[-1])
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
df.to_csv('breast-cancer-wisconsin.csv', index=False)
```
